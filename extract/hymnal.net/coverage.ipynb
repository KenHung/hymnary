```python
import pandas as pd
```

```python
hym = pd.read_csv('../hymnary.csv')
```

```python
matches_ch = pd.read_csv('matches.csv')
```

```python
best_matches = pd.concat(
    [
        hym.query("true_match == True")[
            ["id", "name_ch", "name", "match_title", "match_url"]
        ],
        matches_ch.query("true_match == True")[
            ["id", "name_ch", "match_title_zh", "match_url"]
        ].rename(columns={"match_title_zh": "match_title"}),
    ]
)
```

```python
best_matches.to_csv('best_matches.csv', index=False)
```
