```python
from pathlib import Path
from mido import MidiFile, Message
```

```python
rest_times = {}
for file in Path("alvinwong").iterdir():
    mid = MidiFile(file)
    t = [0]
    for msg in mid.tracks[0]:
        if msg.type == 'time_signature':
            if msg.time > 0:
                t.append(msg.time)

    if max(t) > 200:
        rest_times[file] = max(t)
```

```python
def add_rest_notes(file, time):
    # add rest notes (as dummy middle C)
    mid = MidiFile(file)
    mid.tracks[0].insert(-1, Message("note_on", note=60, velocity=0, time=time))
    mid.save(file)
```

```python
for file, time in rest_times.items():
    file = Path("alvinwong_piano_ai").joinpath(file.name)
    add_rest_notes(file, time)
```

```python
add_rest_notes("alvinwong_piano_ai/004.當讚美聖父.mid", 455)
```
