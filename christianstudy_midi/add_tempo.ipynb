```python
import os
import shutil
import mido

# Set the directory containing the MIDI files
directory = 'temp'

# Initialize counters for the number of invalid files and files without tempo set
invalid_files_count = 0
no_tempo_files_count = 0

# Get the list of MIDI files in the directory, sorted in alphabetical order
midi_files = sorted([f for f in os.listdir(directory) if f.endswith('.mid')])

# Loop through each MIDI file sequentially
for filename in midi_files:
    try:
        # Open the MIDI file
        midi_file = mido.MidiFile(os.path.join(directory, filename))

        # Check if a tempo message has been set
        tempo_set = False
        for track in midi_file.tracks:
            for message in track:
                if message.type == 'set_tempo':
                    tempo_set = True
                    break
            if tempo_set:
                break

        if not tempo_set:
            print(f"No tempo set in {filename}.")
            # Add a default tempo of 500000
            midi_file.tracks[0].insert(0, mido.MetaMessage('set_tempo', tempo=500000))
            # Save the modified file
            midi_file.save(os.path.join(directory, filename))
            no_tempo_files_count += 1
        
    except Exception as e:
        #print(f"Error opening {filename}:{e}")
        invalid_files_count += 1

# Print the count of invalid files and files without tempo set
print(f"Finished processing files. {invalid_files_count} invalid MIDI files.")
print(f"{no_tempo_files_count} MIDI files do not have a tempo set in {directory}.")
```
